1. **List**

  - A List is a **data structure** that can contain more than one value at-a-time.

  - A value in a list is generally referred to as an **item** or an **element**.

  - A List is created using the square brackets `[]`. Each item in a Python list is separated by a comma.

  - List stores Multiple data types values

  - List allows duplicates

  - List is a ordered and mutable in nature

  - Index-Based Access

````python
# Creating a list
l_1 = ["Orange",106.78,2,True,[45,6,7.8,False],2,106.78]
print(l_1)
print(type(l_1))
````

````python
# Accessing elements
# Index(position/rollno for an item) method
# Syntax: list_name[index_no]
# 1. Positive Index (0 to N-1) (Left to Right)
# 2. Negative Index (-1 to -N)

# 1. Positive Index (0 to N-1) (Default)
#          0         1  2  3         4          5   6
l_1 = ["Orange",106.78,2,True,[45,6,7.8,False,[43,"Apple"]],2,106.78] # Nested List
#          -7    -6    -5  -4     -3          -2     -1
# 2. Negative Index (-1 to -N)

print(l_1[-1])
print(l_1[1])
print(l_1[-6])
print(l_1[4][2])
print(l_1[4][4][1])
````

**List Operations ✈**

* `len()` - no of element in the list
* `index()`-  gives the position of the particular element
* `count()` - count the number of times an element appears in the list.
* `sort()` - arranges the values in the asc/desc order
* `append()` - add the element at the end of the list
* `pop()` - delete the values based on the index position
* `remove()` - delete the element based on the values
* `max() `- maximum value of the list
* `min()` - minimum values of the list
* `sum()` - adds all the elements of the list
* `insert()` - add the element at the specific index
* del -  remove items from a list by specifying their index or a slice of indices or whole list
* Checking membership (`in` , `not in`) - check whether an item exists in a Python list or not and return boolean value output.
* List Slicing

# **Set**

  - A Set is a **data structure** that can contain more than one value at-a-time.

  - A Set is created using the curly brackets `{}`.

  - Set stores Multiple data types values

  - Set doesnt allows duplicates

  - Set is an unordered and mutable in nature

  - Data Access is not Possible

````python
# Creating Set
s = {"GUVI",23,67.98,False,67.98,}
print(s)
print(type(s))
````

````python
# Creating Set
s = {"GUVI",23,67.98,[2,4,89],False,67.98}
print(s)
print(type(s))
````

Set Operations ✈

 - `len()` - no of element in the set
 - `sorted() `- arranges the values in the asc/desc order
 - `add()` - add the element at the end of the set
 - `discard()` - delete the element based on the values
 - `remove()` - delete the element based on the values
 - `max()`- maximum value of the set
 - `min()` - minimum values of the set
 - `sum()` - adds all the elements of the set
 - `clear()` - remove all elements from a set
 - Checking membership (`in` , `not in`) - check whether an item exists in a Python set or not and return boolean value output.
 - Union (`|`) / `union ()`: Combines unique elements from both sets.
 - Intersection (`&`): Finds common elements between sets.
 - Difference (`-`): Finds elements in the first set but not in the second.

````python
s1 = {2,5,7,2,4,5,5,10,23,0}
# sorted()
print(sorted(s1))

# sorted()--> sorted(tuple_name,reverse = True)--> des
print(sorted(s1,reverse = True))

# max()
print(max(s1))
# min()
print(min(s1))
# sum()
print(sum(s1))
````

````python
s1 = {"GUVI",23,67.98,False,67.98}
# len()
print(len(s1))

# add()-->set.add(item)
s1.add(10)
print(s1)

# discard()-->set.discard(item)
s1.discard(67.98)
print(s1)

# remove()-->set.remove(item)
s1.remove(23)
print(s1)

# clear()-->set.clear() --> structure/Identity will remain
s1.clear()
print(s1)
````

````python
s1={1,2,3,4}
s2={2,4,5,6}

# union() or  |
print(s1 | s2)

# & (Intersection)
print(s1&s2)

# - (Difference)
print(s1-s2)
print(s2-s1)
````

# **Dictionary**

  - A Dictionary is a **data structure** that can contain more than one value at-a-time.

  - A Dictionary is a collection of key-value pairs** where each key-value pair collectively constitutes an item stored in it.

  - A Dictionary is created using the square brackets `{}`.
  
  - A key and its corresponding value are mapped using the colon (`:`) sign. Each key value pair (item) in a dictionary is separated by a comma.

  - Dictionary stores Multiple data types values

  - Key-Based Access

  - Dictionary is an unordered

  - For Example:

|Key         |Value                  |
|------------|-----------------------|
|Empid       |1,2,3                 |
|Name        |"Shwetha","Geetha","Seetha"               |
|Age         |28,30,32|
|Place       |"Karnataka","Hyderbad","Chennai"            |



````python
# Creating Dictionary
d = {"even":2,"Odd":3}
print(d)
print(type(d))
````

````python
d1 = {"Empid":[1,2,3],"Name":["Shwetha","Geetha","Seetha"],"Age":[28,30,32],
      "Place":["Karnataka","Hyderbad","Chennai"]}
print(d1)
````

````python
# Accessing Elements
# get the geethas empid,name,age,place
print(d1["Name"][1],d1["Empid"][1],d1["Age"][1],d1["Place"][1])
````

**Dictionary Operations ✈**

* `len()` - no of element in the Dictionary
* `keys()` - Returns a view of all values in the Dictionary
* `values()` - Returns a view of all keys in the Dictionary
* `items()` - Returns a view of all key-value pairs as tuples
* `update()` - adding multiple key-value pairs
* `get()` - Retrieves the value associated with key
* `pop()` - Removes the key
* `popitem()` - Remove the last item from the dictionary
* `max() `- maximum value of the values/keys
* `min()` - minimum values of the values/keys
* `sum()` - adds all the elements of the values/keys
* del -  remove items from a list by specifying their index or a slice of indices or whole list
* `clear()` - remove all elements from a Dictionary
* `copy()` - Creates a shallow copy of the dictionary
* Checking membership (`in` , `not in`) - check whether an item exists in a Python list or not and return boolean value output.

````python
#len()
print(len(d1))

#keys()-->dict.keys()
print(d1.keys())

#values()-->dict.values()
print(d1.values())

#items()-->dict.items()
print(d1.items())
````

````python
#get()-->dict.get(key)
print(d1.get("Name"))
print()

#pop()-->dict.pop(key)
d1.pop("Age")
print(d1)
print()

#popitem()-->dict.popitem()
d1.popitem()
print(d1)
````

````python
#update()--> d1.update({'Place': ['Karnataka', 'Hyderbad', 'Chennai']})
d1.update({'Place': ['Karnataka', 'Hyderbad', 'Chennai']})
print(d1)

#Checking membership (`in` , `not in`)
print("Empid" in d1)
print("Age" in d1)

print("Empid" not in d1)
print("Age" not in d1)
````

````python
# copy()
d2 = d1.copy()
print(d2)

# del --> eliminate the Structure/Identity of dict
del d1
print(d1)
````

````python
# clear() --> Structure/Identity of dict will remain
d2.clear()
print(d2)
print(type(d2))
````

# **The Conditional Statements**

-  To Compare two or more values, we use conditional statements.

-  They are of two types of Conditional Statements:

    1. `if-else` statement

    2. `if-elif-else` statement

**1. The `if-else` statement:**

-    The `if-else` statement tells us that **if** a condition is **true**, it will execute a block of statements **else** it will execute another block of statements.



> <center>
<img src="https://student-datasets-bucket.s3.ap-south-1.amazonaws.com/images/if-else-block.png" width=400></center>




> **Syntax of `if-else`:**

> ```python
if condition:
  statement
  statement
else:
  statement
  statement
```

> The statements under `if` block will be executed if the condition is `True`. If the condition is `False`, the statements under the `else` block will be executed.

**1a. Infinite `while` Loop**

- An infinite `while` loop is a never-ending `while` loop.

- **the condition specified in an infinite `while` loop is always true.** To stop an infinite `while` loop, you have to manually stop it by interrupting the code execution.

For example, the `while True` is a never-ending loop. Because the condition `True` is always true.

e.g:
```python
while True:
  //statements
```

**Note:** After interrupting the code execution, Python will throw a `KeyboardInterrupt` error. Ignore it! It's nothing to worry about.

````python
# Infinite while Loop
counter = 1
while True:
  counter = counter * 1
  print(counter)
````

1b. The `break` Keyword

- There's another way to stop a loop whether it is an infinite loop or a finite loop. A finite loop executes for a fixed number of times. To stop a loop, use the `break` keyword.

````python
#Ex:1 break Keyword
counter = 1
while True:
  counter = counter * 2
  print(counter)
  if counter == 8 :
    break
````

````python
#Ex:1 break Keyword
counter = 1
while True:
  counter = counter * 2
  if counter == 8 :
    break
print(counter)
````

2. The `for` Loop


- The `for` loop is used to repeat or iterate a set of statements for a fixed number of times.

- **For e.g:** If we want to print the first `5` natural numbers using `for` loop.

- First, we will use a ready-made function called `range()` inside `for` loop which generates consecutive natural numbers between two specific numbers.

```python
for i in range(1, 6):
  print(i)
```

<img src="https://curriculum.whitehatjr.com/APT+Asset/Untitled-1.gif" width=600/>

**The range function:**

- It is used to generate a range of numbers.

- **Syntax:** `range(start, end)`

Where,

- `start` (optional):  An integer that specifies from which position to start. Default is 0.
- `end` (required): An integer that specifies at which position to stop. This number is not included in the range.

**For e.g:**
1. `range(0,6)` $\Rightarrow$ Generates numbers from 0 to 5.
2.`range(6)` $\Rightarrow$ Generates numbers from 0 to 5.
3. `range(1,11)` $\Rightarrow$ Generates numbers from 1 to 10.
4. `range(11)` $\Rightarrow$ Generates numbers from 0 to 10.

````python
# EX:1 for Loop
for i in range(7):
  print(i)
````

````python
# EX:1 for Loop
for i in range(0,7):
  print("GUVI")
````

````python
# EX:1 for Loop
for i in range(1,8):
  print(i,"GUVI")
````

````python
# EX:2 for Loop
# Even
for i in range(0,11):
  if i%2==0:
    print(i)
````

````python
# EX:3 for Loop
for j in range(0,11,2):
  print(j)
````

````python
for k in range(1,11,2):
  print(k)
````

````python
# print Output
# 9
# 4
# 7
# 2
# 6
a = [9,4,7,2,6]
for i in range(len(a)):
  print(a[i])
````

2.   **User Defined Functions**


*   A function is like a machine that takes some input (not always needed) and gives some output.

*   For example, an ATM is a function that takes your debit card and you’re PIN as inputs and return the cash as an output.

<center>
<img src="https://s3-whjr-v2-prod-bucket.whjr.online/11a4d5fa-317e-4cde-840c-1e84f2340ebb.gif"/></center>

The syntax for creating a user-defined function is

`def function_name(input1, input2, ..., inputN):`

**Note**:

1. The function name must begin with a lowercase letter.

**How to create user defined Functions:**

1.	**Define the Function:**
        -  Choose a name for your function and define it using def.

        `def function_name(input1, input2, ..., inputN):
        `

2.  **Add Code Inside the Function:**

    -  Write the code that you want the function to execute. This can be anything you want!
    ```
          def function_name(input1, input2, ..., inputN):
              statements
```

3.  **Call the Function:**

    - To run the code inside the function, you need to "call" or "invoke" the function.

     ```
function_name(input1, input2, ..., inputN)
```






      




**Categories of User-Defined Functions:**

1.	***Function which takes input and returns output:***

    •	The `return` keyword is to exit a function and return a value.

    • A `return` keyword consists of the return keyword followed by an optional return value.
